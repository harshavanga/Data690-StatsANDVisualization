#### We have special characters and text in the data columns. We will modify this and change it to numerical. 

```python
def Census_year(a):
  a = a.replace('*', '')
  return int(a)

def Total_population(a):
  a = a.replace(',', '')
  return int(a)

def Census_cost(a):
  if 'Billion' in a:
    a = a.replace('$', '')
    a = a.replace(',', '')
    a = a.replace(' Billion', '')
    a = float(a)
    return a*(10**9)
  else:
    a = a.replace('$', '')
    a = a.replace(',', '')
    return float(a)

def Average_cost_per_person(a):
  if ' cents' in a:
    a = a.replace('$', '')
    a = a.replace(' cents', '')
    a = float(a)
    a = a/100
    return a
  else:
    a = a.replace('$', '')
    return float(a)
```

#### Now we will change the columns based on the defined functions.

```python
df['Census Year Clean'] = df['Census Year'].apply(Census_year)
df['Total Population Clean'] = df['Total Population'].apply(Total_population)
df['Census Cost Clean'] = df['Census Cost'].apply(Census_cost)
df['Average Cost Per Person Clean'] = df['Average Cost Per Person'].apply(Average_cost_per_person)
```

```python
df
```

```python
df.drop(['Census Year', 'Total Population', 'Census Cost', 'Average Cost Per Person'],axis = 1,  inplace = True)
```
