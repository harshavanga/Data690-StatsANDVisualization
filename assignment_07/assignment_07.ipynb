```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv')
```

```python
df.shape
```

```python
df.sample(10)
```

```python
df['indicator'].unique()
```

```python
df['Country Name'].unique()
```

Filtering data based on the chosen indicator which is population

```python
df=df[df['indicator']=='SP.POP.TOTL']
df
```

Filtering data for the 5 chosen countries

```python
countries= ['Italy', 'Singapore', 'China', 'United States', 'India']
new_df=df[df['Country Name'].isin(countries)]
new_df
```

```python
df_20=new_df[new_df['Year']==2020]
df_20
```

```python
fig = px.bar(
    df_20, 
    x="Country Name", 
    y="value", 
    title="2020 Population of China, India, Italy, Singapore and United States", 
    color="Country Name", 
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.pie(
    df_20, 
    names="Country Name", 
    values="value", 
    title="2020 Population of China, India, Italy, Singapore and United States", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.box(df_20, x="Country Name", y="value", title='China, India, Italy, Singapore and United States')
fig.show()
```

### Line plot to show change in population for the chosen countries

```python
fig = px.line(new_df, x="Year", y="value", color="Country Name", title='China, India, Italy, Singapore and United States')
fig.show()
```

### Comaring the average population in the different regions

```python
df.Region.unique()
```

```python
df_20 = df_20[df_20.Year == 2020].groupby('Region').value.mean()
df_20
```

```python
fig = px.bar(df_20,x = df_20.index, y = 'value', title = '2020 Life Expectancy of the chosen countries', color = df_20.index)
fig.update_layout(showlegend = False)
fig.show()
```

```python
fig = px.pie(df_20, values= 'value', names=df_20.index, title='Pie chart for the five countries 2020 population')
fig.update_traces(textinfo='value+label')
fig.show()
```

```python
fig = px.box(df_20,  x = df_20.index ,y="value",title = '2020 population of the chosen countries', color = df_20.index)
fig.show()
```

```python
df_regions = new_df.groupby(['Region','Year'])['value'].mean()
df_regions
```

```python
df_regions = df_regions.reset_index()
df_regions
```

```python
fig = px.line(df_regions, x='Year', y = 'value', color='Region')
fig.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv')
df_scatter = df[df['Year'] == 2020]
df_scatter
```

```python
df_scatter = df_scatter.pivot(index = 'Country Name',columns='indicator', values='value')
df_scatter
```

```python
fig = px.scatter(df_scatter, x='NY.GDP.PCAP.PP.CD', y='SP.DYN.LE00.IN', color =df_scatter.index, labels={
                     "NY.GDP.PCAP.PP.CD": "GDP for year 2020",
                     "SP.DYN.LE00.IN": "Life expectancy in 2020"},
                title="GDP vs Life expectancy in 2020")

fig.show()
```
