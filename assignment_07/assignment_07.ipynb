```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv')
```

```python
df['indicator'].unique()
```

```python
df['Country Name'].unique()
```

Filtering data based on the chosen indicator which is population

```python
df=df[df['indicator']=='SP.POP.TOTL']
df
```

Filtering data for the 5 chosen countries

```python
countries= ['Italy', 'Singapore', 'China', 'United States', 'India']
new_df=df[df['Country Name'].isin(countries)]
new_df
```

```python
df_20=new_df[new_df['Year']==2020]
df_20
```

```python
fig = px.bar(
    df_20, 
    x="Country Name", 
    y="value", 
    title="2020 Population of China, India, Italy, Singapore and United States", 
    color="Country Name", 
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.pie(
    df_20, 
    names="Country Name", 
    values="value", 
    title="2020 Population of China, India, Italy, Singapore and United States", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.box(df_20, x="Country Name", y="value", title='China, India, Italy, Singapore and United States')
fig.show()
```

### Line plot to show change in population for the chosen countries

```python
fig = px.line(new_df, x="Year", y="value", color="Country Name", title='China, India, Italy, Singapore and United States')
fig.show()
```
